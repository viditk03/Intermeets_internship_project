```python
import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline
sns.set(color_codes=True)
```

```python
df = pd.read_excel("/content/AirQualityUCI.xlsx")
df.head(5)
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
df = df.drop_duplicates()
df.head(5)
```

```python
df.count()
```

```python
print(df.isnull().sum())
```

```python
# Plot histogram of CO(GT) levels
plt.figure(figsize=(12, 6))
sns.histplot(df['CO(GT)'], bins=30, kde=True)
plt.title('Histogram of CO(GT) Levels')
plt.xlabel('CO(GT)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot time series of CO(GT) levels
plt.figure(figsize=(12, 6))
sns.lineplot(x=df.index, y='CO(GT)', data=df)
plt.title('Time Series of CO(GT) Levels')
plt.xlabel('DateTime')
plt.ylabel('CO(GT)')
plt.show()
```

```python
# Plot histogram of NO2(GT) levels
plt.figure(figsize=(12, 6))
sns.histplot(df['NO2(GT)'], bins=30, kde=True)
plt.title('Histogram of NO2(GT) Levels')
plt.xlabel('NO2(GT)')
plt.ylabel('Frequency')
plt.show()
```

```python

# Plot histogram of PT08.S2(NMHC) levels
plt.figure(figsize=(12, 6))
sns.histplot(df['PT08.S2(NMHC)'], bins=30, kde=True)
plt.title('Histogram of PT08.S2(NMHC) Levels')
plt.xlabel('PT08.S2(NMHC)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot histogram of C6H6(GT)levels
plt.figure(figsize=(12, 6))
sns.histplot(df['C6H6(GT)'], bins=30, kde=True)
plt.title('Histogram of C6H6(GT) Levels')
plt.xlabel('NC6H6(GT)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot histogram of NOx(GT)levels
plt.figure(figsize=(12, 6))
sns.histplot(df['NOx(GT)'], bins=30, kde=True)
plt.title('Histogram of NOx(GT) Levels')
plt.xlabel('NOx(GT)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot between CO and NOx
plt.figure(figsize=(10, 6))
plt.scatter(df['CO(GT)'], df['NOx(GT)'], alpha=0.6, edgecolors='w', linewidth=0.5)
plt.xlabel('CO (GT)')
plt.ylabel('NOx (GT)')
plt.title('CO vs NOx')
plt.show()
```

```python
# Scatter plot between CO and T
plt.figure(figsize=(10, 6))
plt.scatter(df['CO(GT)'], df['T'], alpha=0.6, edgecolors='w', linewidth=0.5)
plt.xlabel('CO (GT)')
plt.ylabel('T')
plt.title('CO vs T')
plt.show()
```

```python
# Scatter plot between PT08.S1(CO) and PT08.S5(O3)
plt.figure(figsize=(10, 6))
plt.scatter(df['PT08.S1(CO)'], df['PT08.S5(O3)'], alpha=0.6, edgecolors='w', linewidth=0.5)
plt.xlabel('PT08.S1(CO')
plt.ylabel('PT08.S5(O3)')
plt.title('PT08.S1(CO) vs PT08.S5(O3)')
plt.show()
```

```python
# Scatter plot between NO2(GT) and NOx(GT)
plt.figure(figsize=(10, 6))
plt.scatter(df['NO2(GT)'], df['NOx(GT)'], alpha=0.6, edgecolors='w', linewidth=0.5)
plt.xlabel('NO2(GT)')
plt.ylabel('NOx(GT)')
plt.title('NO2(GT) vs NOx(GT)')
plt.show()
```

```python
# Scatter plot between NMHC(GT) and C6H6(GT)
plt.figure(figsize=(10, 6))
plt.scatter(df['NMHC(GT)'], df['C6H6(GT)'], alpha=0.6, edgecolors='w', linewidth=0.5)
plt.xlabel('NMHC(GT)')
plt.ylabel('C6H6(GT)')
plt.title('NMHC(GT) vs C6H6(GT)')
plt.show()
```

```python
# Select only numeric columns for correlation matrix
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()

# Heatmap of the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
